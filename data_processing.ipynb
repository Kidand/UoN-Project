```python
import os
import shutil
```

```python
# 创建四个文件夹
folders = ['dataset/3T-Achieva', 'dataset/3T-Ingenia', 'dataset/1.5T-Phillips', 'dataset/1.5T-GE']
for folder in folders:
    if not os.path.exists(folder):
        os.makedirs(folder)
```

```python
# 扫描 dataset 文件夹
def devide_files(dataset_folder):
    files = os.listdir(dataset_folder)
    os.makedirs('dataset/3T-Achieva/' + "imagesTr")
    os.makedirs('dataset/3T-Ingenia/' + "imagesTr")
    os.makedirs('dataset/1.5T-Phillips/' + "imagesTr")
    os.makedirs('dataset/1.5T-GE/' + "imagesTr")
    not_copied_files = []

    for file in files:
        src = os.path.join(dataset_folder, file)
        if not os.path.isfile(src):
            continue

        dst_folder = None

        # 检查文件名并将其复制到相应的文件夹
        if any(keyword in file for keyword in ['AlgiNG', 'AN', 'KIWI', 'MAURTEN', 'Maurten', 'MOLID', 'Nopal']):
            dst_folder = 'dataset/3T-Achieva/' + "imagesTr"
        elif any(keyword in file for keyword in ['GIFT', 'RECLAIM']):
            dst_folder = 'dataset/3T-Ingenia/' + "imagesTr"
        elif any(keyword in file for keyword in ['FABS', 'LETIS', 'Letis', 'MIC']):
            dst_folder = 'dataset/1.5T-Phillips/' + "imagesTr"
        elif any(keyword in file for keyword in ['EFIGI', 'FDA', 'MARC']):
            dst_folder = 'dataset/1.5T-GE/' + "imagesTr"

        if dst_folder is not None:
            dst = os.path.join(dst_folder, file)
            print(dst)
            print(src)
            shutil.copy(src, dst)
        else:
            not_copied_files.append(file)
    return not_copied_files
```

```python
# 打印未被复制的文件
def print_not_copied_files(not_copied_files):
    if not_copied_files:
        print("以下文件未被复制：")
        for file in not_copied_files:
            print(file)
    else:
        print("所有文件已被复制。")
```

```python
not_copied_files = devide_files("images")
print_not_copied_files(not_copied_files)
```

```python
def build_mapping(folders):
    mapping = {}

    for folder in folders:
        folder_i = folder + "/imagesTr"
        for file in os.listdir(folder_i):
            if os.path.isfile(os.path.join(folder_i, file)):
                key = file[:5]
                mapping[key] = folder

    return mapping


folders = ['dataset/3T-Achieva', 'dataset/3T-Ingenia', 'dataset/1.5T-Phillips', 'dataset/1.5T-GE']
mapping = build_mapping(folders)
# 根据 key 排序
mapping = dict(sorted(mapping.items(), key=lambda item: item[0]))
mapping
```

```python
# 打印每个文件夹中的文件数量
for folder in folders:
    print(f"{folder} 文件夹中的文件数量：{len(os.listdir(folder + '/imagesTr'))}")
```

```python
def process_new_files(new_folder, mapping):
    files = os.listdir(new_folder)

    for file in files:
        src = os.path.join(new_folder, file)
        if not os.path.isfile(src):
            continue
        key = file[:5]
        target_folder = mapping.get(key, None) + "/" + "labelsTr"

        if target_folder is not None:
            # 创建目标文件夹（如果不存在）
            if not os.path.exists(target_folder):
                os.makedirs(target_folder)

            dst = os.path.join(target_folder, file)
            shutil.copy(src, dst)
            print(f"文件 {file} 已复制到 {target_folder} 文件夹中。")
        else:
            not_copied_files.append(file)

    if not_copied_files:
        print("以下文件未被复制：")
        for file in not_copied_files:
            print(file)
    else:
        print("所有文件已被复制。")

# 使用示例
new_folder = "masks"
process_new_files(new_folder, mapping)
```

```python
import nibabel as nib

def rename_files(folder):
    # 定义子文件夹路径
    image_folder = os.path.join(folder, 'imagesTr')
    label_folder = os.path.join(folder, 'labelsTr')

    # 获取文件名列表
    image_files = sorted(os.listdir(image_folder))
    label_files = sorted(os.listdir(label_folder))

    # 为 imagesTr 中的文件重命名
    for i, image_file in enumerate(image_files):
        original_image_file = os.path.join(image_folder, image_file)
        new_image_file = os.path.join(image_folder, f'data_{str(i+1).zfill(3)}_0000.nii')
        # 复制并重命名文件
        shutil.copy2(original_image_file, new_image_file)
        # 删除原文件
        os.remove(original_image_file)

        # 将文件从 .nii 转换为 .nii.gz
        img = nib.load(new_image_file)
        nib.save(img, new_image_file[:-4] + '.nii.gz')
        os.remove(new_image_file)

        # 打印重命名日志
        print(f"Renamed {original_image_file} to {new_image_file[:-4] + '.nii.gz'}")

    # 为 labelsTr 中的文件重命名
    for i, label_file in enumerate(label_files):
        original_label_file = os.path.join(label_folder, label_file)
        new_label_file = os.path.join(label_folder, f'data_{str(i+1).zfill(3)}.nii')
        # 复制并重命名文件
        shutil.copy2(original_label_file, new_label_file)
        # 删除原文件
        os.remove(original_label_file)

        # 将文件从 .nii 转换为 .nii.gz
        img = nib.load(new_label_file)
        nib.save(img, new_label_file[:-4] + '.nii.gz')
        os.remove(new_label_file)

        # 打印重命名日志
        print(f"Renamed {original_label_file} to {new_label_file[:-4] + '.nii.gz'}")




# 使用示例
rename_files("dataset/3T-Achieva")
rename_files("dataset/3T-Ingenia")
rename_files("dataset/1.5T-Phillips")
rename_files("dataset/1.5T-GE")
```

```python
import os
import json

def generate_dataset_json(folder_name, task_id):
    # Count the number of files in the imagesTr subfolder
    num_training = len(os.listdir(folder_name + '/labelsTr'))

    # Define the data for the JSON file
    data = {
        "task_id": task_id,
        "file_ending": ".nii.gz",
        "name": folder_name,
        "description": f"bowel segmentation {folder_name}",
        "tensorImageSize": "4D",
        "modality": {
            "0": "T2"
        },
        "channel_names": {
            "0": "MRI"
        },
        "labels": {
            "background": 0,
            "bowel": 1
        },
        "numTraining": num_training
    }

    # Set the path for the JSON file
    json_file_path = os.path.join(folder_name, "dataset.json")

    # Write the data to the JSON file
    with open(json_file_path, 'w') as json_file:
        json.dump(data, json_file)

    print(f"dataset.json has been created in the folder {folder_name}.")

# Test the function
generate_dataset_json("dataset/1.5T-GE", "002")
generate_dataset_json("dataset/1.5T-Phillips", "003")
generate_dataset_json("dataset/3T-Achieva", "004")
generate_dataset_json("dataset/3T-Ingenia", "005")
```

```python
# process dir
import os

import os

def retain_nii_gz_files(directory):
    # os.walk生成文件夹的文件名，子文件夹，子文件
    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            # 如果文件名不是以.nii.gz结尾，删除该文件
            if not filename.endswith('.nii.gz'):
                file_path = os.path.join(dirpath, filename)
                try:
                    os.remove(file_path)
                except OSError as e:
                    print(f"Error: {file_path} : {e.strerror}")
```
