```python
import nibabel as nib
from nilearn import plotting

# 读取nii文件
nii_file = "dataset/1.5T-GE/imagesTr/data_001_0000.nii.gz"
img = nib.load(nii_file)

# 读取预测结果
prediction = nib.load("dataset/1.5T-GE/predictions/data_001.nii.gz")

# 可视化
display = plotting.plot_anat(img)
display.add_overlay(prediction, cmap='jet')  # 添加分割结果作为覆盖层
plotting.show()
```

```python
from nilearn import plotting, datasets

# 读取nii文件
nii_file = "dataset/1.5T-GE/imagesTr/data_001_0000.nii.gz"
img = nib.load(nii_file)

# 读取预测结果
prediction = nib.load("dataset/1.5T-GE/predictions/data_001.nii.gz")

# 使用nilearn的fsaverage5数据集作为显示的表面
fsaverage = datasets.fetch_surf_fsaverage()

# 可视化
view = plotting.view_img_on_surf(prediction, surf_mesh=fsaverage, bg_map=img, cmap='hot', threshold=0.5)
view.open_in_browser()

```
